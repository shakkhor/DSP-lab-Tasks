## Task-7

```python
def Adder(x1,x2):
    l = len(x1)
    y = []
    for i in range(l):
        y.append(x1[i] + x2[i])
    return y

        
def Multiplier(mul, x):
    l = len(x)
    y = []
    for i in range(l):
        y.append( (x[i]) * mul )
    return y

def Delay(d, x):
    l = len(x)
    y = []
    for i in range(l):
        y.append(x[i])
    for i in range(d):
        y.append(0)
    return y

def Advance(a, x):
    l = len(x)
    y = []
    for i in range(a):
        y.append(0)
    for i in range(l):
        y.append(x[i])
    return y

def Fold(x):
    return x[::-1]

fig = plt.figure(figsize= [16,9])
x1 = [0,0,1,2,3,4]
x2 = [0,0,4,3,2,1]
x  = [1,2,3,4]
d = 2
mul = 5
a = 2

print("Adder: ",Adder(x1,x2))
print("Multiplier: ",Multiplier(mul, x))
print("Delay: ",Delay(d, x))
print("Advance: ",Advance(a, x))
print("Fold: ",Fold(x))
xx = [-1,0,1,2,3,4] 
plt.subplot(3,3,1)
plt.title("X1", fontweight="bold", fontsize= 15)
plt.stem(xx,x1)

plt.subplot(3,3,2)
plt.title("X2", fontweight="bold", fontsize= 15)
plt.stem(xx,x2)

plt.subplot(3,3,3)
plt.title("X1 + X2", fontweight="bold", fontsize= 15)
plt.stem(x1,Adder(x1,x2))

plt.subplot(3,3,4)
plt.title("X1*5", fontweight="bold", fontsize= 15)
plt.stem(x1,Multiplier(mul, x1))

xx = [-1,0,1,2,3,4] 

plt.subplot(3,3,5)
plt.title("X1 Delayed 2 Units", fontweight="bold", fontsize= 15)
plt.stem(xx,Delay(d, x))
xx1 = [1,2,3,4,5,6] 

plt.subplot(3,3,6)
plt.title("X1 Advanced 2 Units", fontweight="bold", fontsize= 15)
plt.stem(xx1,Advance(d, x))

plt.subplot(3,3,7)
xx2 = [-4,-3,-2,-1,0,1]
plt.title("X1 Folded", fontweight="bold", fontsize= 15)
plt.stem(xx2,Fold(x1))

plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=0.3)
```

## Task-8

```python
def movingAverage(values,window):
    weigths = np.repeat(1.0, window)/window
    m = np.convolve(values, weigths, 'valid')
    #print ("m: ",m)
    return m

x = np.arange(-10, 10, 0.25)
x[13] = 1.02
x[14] = 1.05

x[27] = -0.78
x[28] = -0.92
x[30] = -0.75

signal = np.sin(x)
fig = plt.figure(figsize= [20,10])
plt.subplot(2,1,1)       
plt.title("Moving Average Filter", fontweight="bold", fontsize= 20)
plt.plot(movingAverage(signal,5), 'r-', label="Filtered Signal")
plt.plot(signal, 'g', label='Original singal')
plt.legend(prop={'size': 20})
#plt.show()

def medianFilter(data):
    return medfilt(data, kernel_size=5)

plt.subplot(2,1,2)       
plt.title("Median Filter", fontweight="bold", fontsize= 20)
plt.plot(medianFilter(signal), 'r', label="Filtered Signal")
plt.plot(signal, 'g',label='Original singal')
plt.legend(prop={'size': 20})
plt.show()

```

## Task-10

```python
import numpy.fft as DFT ### Getting DFT using fft
N = 100
f=1/10
t = np.linspace(0,100,N)
x = np.cos(2*np.pi*.25*f*t)
y = DFT.fftn(x, norm='ortho')
window = np.zeros(N)
window[:20] = 1
yr = DFT.ifftn(y*window, norm='ortho')

fig = plt.figure(figsize= [16,9])

plt.plot(t, x, '-bx', label= "Original Siganl")

plt.plot(t, yr, 'ro', label =  "Signal After DFT")
plt.legend(prop={'size': 15})
plt.title("Simple Singal DFT", fontweight="bold", fontsize= 20 )
plt.show()
```

## Task-11

```python
import numpy.fft as DFT ### Getting DFT using fft
N = 500
f=1/10
t = np.linspace(0,20,N)
x = np.sin(2*np.pi*f*t) + np.exp(-t/3)*np.cos(2*np.pi*f*t)+ 2*np.cos(2*np.pi*5*f*t) ##Composite signal

y = DFT.fftn(x, norm='ortho')
window = np.zeros(N)
window[:20] = 1
yr = DFT.ifftn(y*window, norm='ortho')

fig = plt.figure(figsize= [16,9])

plt.plot(t, x, '-bx', label= "Original Siganl")

plt.plot(t, yr, 'ro', label =  "Signal After DFT")
plt.legend(prop={'size': 15})
plt.title("Composite Singal DFT", fontweight="bold", fontsize= 20 )
plt.show()
```
