## Using librosa

```python
import numpy, scipy, sklearn, librosa, urllib, IPython.display
import matplotlib.pyplot as plt
# plt.rcParams['figure.figsize'] = (14,4)
import librosa.display
```

```python
plt.figure(figsize=(16, 9))

x, fs = librosa.load('test.wav', duration=10)
librosa.display.waveplot(x, sr=fs, color='k', alpha=0.5)

plt.title("Original Signal", fontweight = "bold", fontsize = 20)
plt.show()
```

```python
plt.figure(figsize=(16, 9))
mfccs = librosa.feature.mfcc(x, sr=fs)
mfccs = sklearn.preprocessing.scale(mfccs, axis=1)
librosa.display.specshow(mfccs, sr=fs, x_axis='time')
plt.title(".wav file to MFCC(using librosa)", fontweight = "bold", fontsize = 20)
plt.show()
```

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

img = cv2.imread('plot.png',0)

dft = cv2.dft(np.float32(img),flags = cv2.DFT_COMPLEX_OUTPUT)
dft_shift = np.fft.fftshift(dft)

magnitude_spectrum = 20*np.log(cv2.magnitude(dft_shift[:,:,0],dft_shift[:,:,1]))

plt.subplot(121),plt.imshow(img, cmap = 'gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(magnitude_spectrum, cmap = 'gray')
plt.title('Magnitude Spectrum'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def shade(imag, percent):
    tinted_imag = imag * (1 - percent)
    return tinted_imag

def vertical_gradient_line(image, reverse=False):
    number_of_columns = image.shape[1]
    if reverse:
        C = np.linspace(1, 0, number_of_columns)
    else:
        C = np.linspace(0, 1, number_of_columns)
    C = np.dstack((C, C, C))
    return C

def horizontal_gradient_line(image, reverse=False):
    number_of_rows, number_of_columns = image.shape[:2]
    C = np.linspace(1, 0, number_of_rows)
    C = C[np.newaxis,:]
    C = np.concatenate((C, C, C)).transpose()
    C = C[:, np.newaxis]
    return C
plt.figure(figsize=(10, 6))
img  = plt.imread('R5.png')
plt.imshow(img)
plt.show()
plt.figure(figsize=(10, 6))
img = plt.imread('R5.png')
windmills = plt.imread('R5.png')
tinted_windmills = shade(windmills, 0.7)


plt.imshow(tinted_windmills)
plt.title("Basic Image Processing(Shade)", fontweight = "bold", fontsize = 20)

plt.show()
```
