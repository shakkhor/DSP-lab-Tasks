## Task-2



```python
x = np.arange(-10, 10, 0.005)
y = np.sin(x)

fig = plt.figure(figsize= [16,9])
plt.plot(x,y ,'*',color = 'K')
plt.title("Sine Wave", fontweight="bold", fontsize= 20)
plt.xlabel("Time", fontsize= 20 )
plt.ylabel("Amplitude", fontsize= 20)
plt.grid(True, which = 'both')
plt.axhline(y=0, color='k')

 

plt.show()
```

## Task-3

```python
x = np.arange(-5, 5, 0.005)
y1 = np.sin(x)
y2 = np.sin(5*x)
y3 = np.sin(0.5*x)
y = y1+y2+y3

fig = plt.figure(figsize= [16,9])



plt.plot(x,y, '*' ,label='Y',)
plt.plot(x,y1, label='Y1')
plt.plot(x,y2, label='Y2')
plt.plot(x,y3, label='Y3')
plt.title("Composite Sine Wave(Y = Y1+Y2+Y3)", fontweight="bold", fontsize= 20)
plt.xlabel("Time", fontsize= 20)
plt.ylabel("Amplitude", fontsize= 20)
plt.grid(True, which = 'both', )
plt.axhline(y=0, color='k')
plt.legend(prop={'size': 20})
 

plt.show()
```

## Task-4

```python
fig = plt.figure(figsize= [16,9])

x = np.arange(-25,25,1)
y = []
#unit impulse
for i in x:
    if i == 0:
        y.append(1)
    else:
        y.append(0)
plt.subplot(2,2,1)      
plt.title("Unit Impulse", fontweight="bold", fontsize= 20)        
plt.stem(x,y)
plt.grid(True, which = 'both')
plt.axhline(y=0, color='r')


#unit step
y1 = []
for i in x:
    if i >= 0:
        y1.append(1)
    else:
        y1.append(0)
plt.subplot(2,2,2)      
plt.title("Unit Step", fontweight="bold", fontsize= 20)        
plt.stem(x,y1)
plt.grid(True, which = 'both')
plt.axhline(y=0, color='r')



#ramp
y2 = []
for i in x:
    if i >= 0:
        y2.append(i)
    else:
        y2.append(0)
plt.subplot(2,2,3)      
plt.title("Ramp", fontweight="bold", fontsize= 20)        
plt.stem(x,y2)
plt.grid(True, which = 'both')
plt.axhline(y=0, color='r')


#Exponential
y3 = []
for i in x:
    if i >= 0:
        y3.append(i*i)
    else:
        y3.append(0)
plt.subplot(2,2,4)      
plt.title("Exponential", fontweight="bold", fontsize= 20)        
plt.stem(x,y3)



plt.grid(True, which = 'both')
plt.axhline(y=0, color='r')
plt.show()


```

## Task-5

```python
def sinWave(a,T,t):
    f = 1/T
    y=a*np.sin(2*np.pi*f*t)
    return y

def SubPloting(nrows,ncols,index,t,y,color,title,xlabel,ylabel,markersize,show):
    plt.subplot(nrows,ncols,index)
    plt.title(title, fontweight="bold", fontsize= 20)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    l=plt.plot(t,y,color)
    plt.grid(True, which = 'both')
    plt.axhline(y=0, color='r')
    if markersize > 0:
        plt.setp(l,markersize=markersize)
    if show==1:
        plt.show()

fig = plt.figure(figsize= [20,15])
T=20
f=1/T
fs=2*f

t=np.arange(-25, 25, 0.01)
y=sinWave(10,20,t)
SubPloting(2,2,1,t,y,'r',"Original Signal",'','',0,0)

# fs = fs-0.05
fs = fs+1
tdif = 1/fs
t=np.arange(-25, 25, tdif)
y=sinWave(5,T,t)
markersize=10
SubPloting(2,2,2,t,y,'g*',"Sampled Signal",'','',markersize,1)
        
    
```
