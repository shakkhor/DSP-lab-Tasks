```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def sinWave(a,T,t):
    f = 1/T
    y=a*np.sin(2*np.pi*f*t)
    return y
```

### Task#02: Create a Sinusoidal Signal y(n) and plot the signal.

```python
t=np.arange(0.0,100,0.01)
y=sinWave(5,10,t)
plt.plot(t,y, 'm')
plt.show()
```

### Task#03: Create 3 Sinusoidal signal y1, y2, y3. then combine them and plot it.

```python
t=np.arange(0.0,100,0.01)

plt.subplot(2,2,1)
y1=sinWave(5,10,t)
plt.plot(t,y1, 'r')

plt.subplot(2,2,2)
y2=sinWave(5,20,t)
plt.plot(t,y2, 'b')

plt.subplot(2,2,3)
y3=sinWave(5,25,t)
plt.plot(t,y3, 'g')

plt.show()

y=y1+y2+y3
plt.plot(t,y, 'o')
plt.show()
```

### Task#04: Plot the basic signals like unit impulse, unit samples, unit step, exponential, unit ramp (discussed in classes) etc.

```python
fig = plt.figure(figsize= [20,10])
###Unit Impulse
t=np.arange(-20,20,1)
y=[]
mid=0
for x in t:
    if x==mid:
        y.append(1)
    else:
        y.append(0)
plt.subplot(2,2,1)      
plt.title("Unit Impulse")
plt.stem(t,y)

###unit Step
t1=np.arange(-20,20,1)
y1=[]
for x in t1:
    if x>=0:
        y1.append(1)     
    else:
        y1.append(0)
plt.subplot(2,2,2)       
plt.title("Unit Step")
plt.stem(t1,y1)

###unit Exponential
t2=np.arange(-20,20,0.1)
y2=[]
for x in t2:
    y2.append(np.exp(x))
plt.subplot(2,2,3)       
plt.title("Exponential")
plt.plot(t2,y2)

###unit ramp
t3=np.arange(-20,20,0.1)
y3=[]
for x in t3:
    if x>=0:
        y3.append(x)
    else:
        y3.append(0)
plt.subplot(2,2,4)       
plt.title("Ramp")
plt.stem(t3,y3, 'c')

plt.show()
```

### Task#05: Sampling of a Signal... take a signal then sample using a sampling rate.

```python
def SubPloting(nrows,ncols,index,t,y,color,title,xlabel,ylabel,markersize,show):
    plt.subplot(nrows,ncols,index)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    l=plt.plot(t,y,color)
    if markersize > 0:
        plt.setp(l,markersize=markersize)
    if show==1:
        plt.show()
```

```python
fig = plt.figure(figsize= [20,10])
T=10
f=1/T
fs=2*f

t=np.arange(0.0, 50, 0.01)
y=sinWave(5,10,t)
SubPloting(2,2,1,t,y,'b',"Main Signal",'','',0,0)

# fs = fs-0.05
fs = fs+1
tdif = 1/fs
t=np.arange(0.0, 50, tdif)
y=sinWave(5,T,t)
markersize=10
SubPloting(2,2,2,t,y,'ro',"After Sampling",'','',markersize,1)
```

### Task#06: Sampling Theorem Experiment.

```python
fig = plt.figure(figsize= [20,10])
T=10
f=1/T
fs=2*f

t=np.arange(0.0, 50, 0.01)
y=sinWave(5,10,t)
SubPloting(2,2,1,t,y,'b',"Main Signal",'Time =>','Amplitue =>',0,0)

# fs = fs-0.05
fs = fs+0.1
tdif = 1/fs
t=np.arange(0.0, 50, tdif)
y=sinWave(5,T,t)
markersize=10
SubPloting(2,2,2,t,y,'ro',"After Sampling",'Time =>','Amplitue =>',markersize,1)
```

### Task#08: Implement Moving average filter, Median Filter, Accumulator etc.

```python
import os
import sys
import seaborn
from scipy.signal import medfilt
```

```python
def movingaverage(values,window):
    weigths = np.repeat(1.0, window)/window
    m = np.convolve(values, weigths, 'valid')
    #print ("m: ",m)
    return m

signal = [1,6,3,13,5,6,7,10,9,4,6,5,7]

plt.subplot(2,2,1)       
plt.title("Moving Average")
plt.plot(movingaverage(signal,3), 'r-')
plt.show()

def median_filter(data):
    return medfilt(data, kernel_size=3)

plt.subplot(2,2,1)       
plt.title("Median Filter")
plt.plot(median_filter(signal), 'r-')
plt.show()
```

### Task#10: program to find the DFT/IDFT

```python
import numpy.fft as DFT ### Getting DFT using fft
N = 100
f=1/10
t = np.linspace(0,20,N)
x = np.sin(2*np.pi*f*t)

y = DFT.fftn(x, norm='ortho')
window = np.zeros(N)
window[:20] = 1
yr = DFT.ifftn(y*window, norm='ortho')

plt.plot(t, x, '-bx')
plt.plot(t, yr, 'ro')
plt.show()
```

### Tasks#11: DFT/IDFT -> Composite Sinusoidal Signal to see the frequencies

```python
import numpy.fft as DFT ### Getting DFT using fft
N = 100
f=1/10
t = np.linspace(0,20,N)
x = np.sin(2*np.pi*f*t) + np.exp(-t/3)*np.cos(2*np.pi*f*t) ##Composite signal

y = DFT.fftn(x, norm='ortho')
window = np.zeros(N)
window[:20] = 1
yr = DFT.ifftn(y*window, norm='ortho')

plt.plot(t, x, '-bx')
plt.plot(t, yr, 'ro')
plt.show()
```
